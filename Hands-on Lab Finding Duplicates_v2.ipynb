## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
duplicates = df[df.duplicated()]

# Display duplicates
print(len(duplicates))
duplicates.head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork.
subset_cols = ['MainBranch', 'Employment', 'RemoteWork']
subset_duplicates = df[df.duplicated(subset=subset_cols, keep=False)]
print(f"\n No Duplicate rows based on {subset_cols}: {subset_duplicates.shape[0]}")
```

```python
identical_cols = subset_duplicates.nunique()==1
identical_columns = identical_cols[identical_cols].index.tolist()
identical_columns
```

```python
# Number of duplicate groups
duplicate_groups_count = subset_duplicates.groupby(subset_cols).ngroups
print(f"Number of duplicate groups: {duplicate_groups_count}\n")

# Initialize dictionary to store identical value count per column
identical_value_counts = {}

# Check each other column
for col in df.columns:
    if col not in subset_cols:
        # For each duplicate group, count unique values for this column
        identical_in_groups = (subset_duplicates.groupby(subset_cols)[col]
                               .nunique()
                               .eq(1)
                               .sum())
        identical_value_counts[col] = identical_in_groups

# Convert to DataFrame for clean display
result_df = pd.DataFrame.from_dict(identical_value_counts, orient='index', columns=['Identical_in_Groups'])
result_df.sort_values(by='Identical_in_Groups', ascending=False, inplace=True)

# Display result
print("Columns frequently containing identical values within duplicate groups:\n")
print(result_df)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
country_counts = subset_duplicates['Country'].value_counts()
plt.figure(figsize=(10, 6))
country_counts.head(10).plot(kind='bar', color='blue')
plt.title('Distribution of Duplicates by Country')
plt.xlabel('Country')
plt.ylabel('Number of Duplicate Rows')
plt.show()

employment_counts = subset_duplicates['Employment'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(employment_counts.head(10), labels=employment_counts.head(10).index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Duplicates by Employment Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
df.drop_duplicates(subset=subset_cols,keep='first', inplace=True)
duplicates = df[df.duplicated()]

print(len(duplicates))
```
