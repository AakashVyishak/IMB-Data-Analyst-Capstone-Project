```python
import pandas as pd
import matplotlib.pyplot as plt

df_databases_want = pd.read_csv("survey_data_updated 5.csv")

database_counts = df_databases_want['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()

top_5_databases = database_counts.nlargest(10).sort_values(ascending=True)
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F',
          '#EDC948', '#B07AA1', '#FF9DA7', '#9C755F', '#BAB0AC']
# Create a figure and axes for the plot
plt.figure(figsize=(6, 4))

# Plot a histogram/bar chart of the top 5 databases
top_5_databases.plot(kind='barh',color=colors)

plt.xlabel('Programing Languages')
plt.ylabel('Number of Respondents')
plt.title('Top 10 Most Used Programing Languages')
plt.xticks(rotation=45, ha='right') # Rotate labels for readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
df_lang_salary = df_databases_want.dropna(subset=['LanguageHaveWorkedWith', 'ConvertedCompYearly']).copy()

# Split the 'LanguageHaveWorkedWith' by ';' and explode into separate rows
df_lang_salary['LanguageList'] = df_lang_salary['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_lang_salary.explode('LanguageList')

# Clean whitespace around languages
df_exploded['LanguageList'] = df_exploded['LanguageList'].str.strip()

# Group by language and calculate average salary
avg_salary_by_language = df_exploded.groupby('LanguageList')['ConvertedCompYearly'].mean().sort_values(ascending=False)

# Get top 10 languages by average salary
top_10_avg_salary = avg_salary_by_language.nlargest(10)

# Plot bar chart
plt.figure(figsize=(8, 5))
top_10_avg_salary.plot(kind='bar', color='#4E79A7')

plt.xlabel('Programming Language')
plt.ylabel('Average Salary (Yearly)')
plt.title('Top 10 Programming Languages by Average Salary')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_jp = pd.read_excel(" job-postings.xlsx")
df_jp = df_jp[['Languages', 'Number of Jobs']]

# Sort the DataFrame in descending order of Number of Jobs
df_jp = df_jp.sort_values(by='Number of Jobs', ascending=False)
plt.figure(figsize=(10, 5))
plt.bar(df_jp['Languages'], df_jp['Number of Jobs'], color='#4E79A7')

plt.xlabel('Languages')
plt.ylabel('No of Jobs Posted')
plt.title('No of Jobs Postings vs Languages')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_ppl = pd.read_csv("popular-languages.csv")
df_ppl = df_ppl[['Language', 'Average Annual Salary']]
df_ppl['Average Annual Salary'] = (
    df_ppl['Average Annual Salary']
    .replace('[\$,]', '', regex=True)    # remove $ or commas
    .replace(',', '', regex=True)        # extra comma removal just in case
)
df_ppl['Average Annual Salary'] = pd.to_numeric(df_ppl['Average Annual Salary'], errors='coerce')

# Sort the DataFrame in descending order of Number of Jobs
df_ppl = df_ppl.sort_values(by='Average Annual Salary', ascending=False)
plt.figure(figsize=(10, 5))
plt.bar(df_ppl['Language'], df_ppl['Average Annual Salary'], color='#4E79A7')

plt.xlabel('Languages')
plt.ylabel('Average Annual Salary')
plt.title('Languages Vs Average Annual Salary')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
